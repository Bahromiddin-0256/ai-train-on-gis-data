# Exploring Sentinel-2 imagery over the Fergana Valley (Uzbekistan)

A starter notebook that shows how to:

1. Search the Planetary Computer STAC API for Sentinel-2 L2A scenes.
2. Visualise the AOI on a map with `folium`.
3. Load a sample tile with `rasterio` and plot an RGB composite.

````python
from gis_train.data.download import search_sentinel2_l2a
from gis_train.utils.geo import BBox

fergana = BBox(70.60, 40.10, 72.90, 41.40)
fergana
````

````python
import folium

centre = [(fergana.lat_min + fergana.lat_max) / 2, (fergana.lon_min + fergana.lon_max) / 2]
m = folium.Map(location=centre, zoom_start=8)
folium.GeoJson(fergana.as_geojson(), name="AOI").add_to(m)
m
````

````python
items = list(
    search_sentinel2_l2a(
        bbox=fergana,
        date_start="2021-07-01",
        date_end="2021-07-31",
        cloud_cover_max=10.0,
        limit=5,
    )
)
for item in items:
    print(item.id, item.properties.get("eo:cloud_cover"))
````
